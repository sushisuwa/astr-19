```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
n = 1000
x = np.random.uniform(low=0,high=1,size=n)

f, ax = plt.subplots(1,1,figsize=(4,4))

ax.hist(x,bins=100,edgecolor='black',alpha=1.0)
ax.set_xlabel('x')
ax.set_ylabel('y')
plt.savefig('s9.pdf',bbox_inches='tight',dpi=600)
```
