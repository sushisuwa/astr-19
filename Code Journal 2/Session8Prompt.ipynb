```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0,1,0.1)
```

```python
def sinx(x):
    return np.sin(x)
```

```python
def cosx(x):
    return np.cos(x)
```

```python
y = sinx(x)
z = cosx(x)
f, axarr = plt.subplots(1,2)

axarr[0].plot(x,y)
axarr[0].set_xlabel('x')
axarr[0].set_ylabel('sin(x)')
axarr[0].set_title(r'$\sin(x)$')

axarr[1].plot(x,z)
axarr[1].set_xlabel('x')
axarr[1].set_ylabel('cos(x)')
axarr[1].set_title(r'$\cos(x)$')

f.subplots_adjust(wspace=0.4)

plt.savefig('s8.pdf',bbox_inches='tight',dpi=600)
```
