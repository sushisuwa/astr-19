```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

Create an array x ranging [0,1] with 100 steps

```python
x = np.arange(0,1,.01)
```

Create a functions that returns exp(x)

```python
def exp(x):
    return np.exp(x)
```

Draw the graph

```python
y = exp(x)
plt.plot(x,y)
plt.xlabel('Time[milliseconds]')
plt.ylabel('Awesomeness')
plt.savefig('exp.jpg',bbox_inches='tight',dpi=600)
```
